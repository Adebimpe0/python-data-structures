# A student Scores Tracker

This section uses a list to store and manage student test scores.

```python
score_list = [65,78,90,55,88,72] #Create a list of student test scores
```

```python
print("Highest score:" , max(score_list)) #Print the highest test scores
```

```python
print("Lowest score:" , min(score_list)) #Print the lowest test scores
```

```python
average_score = sum(score_list) / len(score_list) #Calculate the average _score
```

```python
print("Average score:", average_score) #Print the average_score
```

```python
score_list.append(80) #Add a new score
```

```python
score_list.remove(55) #Remove the failed score
```

```python
print("Final score_list:" , score_list)
```

# Explanation: A list is used because student scorces can change. The code finds the highest and lowest scores, calcuates the average, and updates the list by adding and removing values.

# B. Shopping Cart System

This section demonstrates how a list can be used to manage shopping items.

```python
cart = ["rice", "beans", "oil", "tomatoes"] #create a shopping cart list
```

```python
cart.append("salt")
cart.append("onions") #Add new items to the cart
```

```python
cart.remove("oil") #Remove an item
```

```python
print("Final.cart:", cart)
print("Total items:", len(cart))
```

# Explanation: The shopping cart is stored in a list because items can be added or removed. The code updates the cart and displays the total number of items.

# C. Daily Temperature Records. This section uses a tuple to store daily temperature values.

```python
temperature_tuple = [28, 30, 31, 29, 27] #Store temperatures in a tuple
```

```python
print("Wednesday temperature:", temperature_tuple[2])
```

```python
print("Highest temperature:", max(temperature_tuple))
print("Lowest temperature:" , min(temperature_tuple)) #Print highest and lowest temperatures
```

# A tuple is used because temperature records should not be changed. The code accesses a specific day and finds the highest and lowest values.

# Why a Tuple is better Than a List.A tuple is immutable, meaning its values cannot be changed after creation.This makes it suitable for storing fixed data such as weather records.
